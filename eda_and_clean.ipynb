```python
print("\n--- 2. Análisis de Variables Numéricas ---")
numerical_cols_eda = df_final.select_dtypes(include=np.number).columns.tolist()
if 'condition_encoded' in numerical_cols_eda:
    numerical_cols_eda.remove('condition_encoded')

numerical_cols_for_dist_eda = ['base_price', 'price', 'initial_quantity', 'sold_quantity', 'available_quantity']
numerical_cols_for_dist_eda = [col for col in numerical_cols_for_dist_eda if col in df_final.columns and col in numerical_cols_eda]


if numerical_cols_for_dist_eda:
    print("\nEstadísticas Descriptivas de Variables Numéricas Seleccionadas:")
    print(df_final[numerical_cols_for_dist_eda].describe())
    for col in numerical_cols_for_dist_eda:
        plt.figure(figsize=(12, 5))
        plt.subplot(1, 2, 1)
        sns.histplot(df_final[col].dropna(), kde=True, color='skyblue', bins=50)
        plt.title(f'Distribución de {col}')
        plt.xlabel(col)
        plt.ylabel('Frecuencia')
        if 'condition' in df_final.columns:
            plt.subplot(1, 2, 2)
            sns.boxplot(x='condition', y=col, data=df_final, hue='condition', palette='pastel', legend=False)
            plt.title(f'{col} vs. Condition')
            plt.xlabel('Condición del Producto')
            plt.ylabel(col)
        plt.tight_layout()
        plt.show()
        if col in ['price', 'base_price']:
            plt.figure(figsize=(12, 5))
            plt.subplot(1, 2, 1)
            sns.histplot(np.log1p(df_final[col].fillna(0)), kde=True, color='lightcoral', bins=50)
            plt.title(f'Distribución de log({col}+1)')
            plt.xlabel(f'log({col}+1)')
            plt.ylabel('Frecuencia')
            if 'condition' in df_final.columns:
                plt.subplot(1, 2, 2)
                df_temp_log = df_final.copy()
                df_temp_log[f'log_{col}'] = np.log1p(df_temp_log[col].fillna(0))
                sns.boxplot(x='condition', y=f'log_{col}', data=df_temp_log, hue='condition', palette='pastel', legend=False)
                plt.title(f'log({col}+1) vs. Condition')
                plt.xlabel('Condición del Producto')
                plt.ylabel(f'log({col}+1)')
            plt.tight_layout()
            plt.show()
else:
    print("No se encontraron columnas numéricas seleccionadas para distribuciones en EDA.")
```

```python
print("\n--- 3. Análisis de Variables Categóricas ---")
categorical_cols_eda = df_final.select_dtypes(include='object').columns.tolist()
cols_to_exclude_from_cat_eda = ['condition', 'title', 'warranty', 'tags', 'shipping_methods', 'non_mercado_pago_payment_methods', 'date_created']
categorical_cols_eda = [col for col in categorical_cols_eda if col not in cols_to_exclude_from_cat_eda]

ids_as_categories_eda = ['listing_type_id', 'id_category', 'buying_mode']
for id_col in ids_as_categories_eda:
    if id_col in df_final.columns and id_col not in categorical_cols_eda:
        if df_final[id_col].dtype == 'object' or (np.issubdtype(df_final[id_col].dtype, np.number) and df_final[id_col].nunique() < 20):
             categorical_cols_eda.append(id_col)

if 'seller_id' in df_final.columns and 'seller_id' not in categorical_cols_eda:
    if df_final['seller_id'].nunique() > 20:
        categorical_cols_eda.append('seller_id')

for col in categorical_cols_eda:
    if col in df_final.columns:
        print(f"\nAnálisis de la columna categórica: {col}")
        top_n = 10
        print(f"Top {top_n} categorías más frecuentes (incluyendo NaNs como categoría si los tiene):")
        value_counts_series = df_final[col].value_counts(normalize=True, dropna=False)
        print(value_counts_series.head(top_n) * 100)

        plt.figure(figsize=(12, 5))
        plt.subplot(1, 2, 1)

        order_main_temp = df_final[col].value_counts(dropna=False).iloc[:top_n]
        if pd.isna(order_main_temp.index).any():
             df_temp_countplot = df_final.copy()
             df_temp_countplot[col] = df_temp_countplot[col].fillna("NaN_placeholder_EDA")
             order_main = df_temp_countplot[col].value_counts().iloc[:top_n].index
             sns.countplot(y=col, data=df_temp_countplot, order=order_main, hue=col, palette='viridis', legend=False, dodge=False)
        else:
             order_main = df_final[col].value_counts().iloc[:top_n].index
             sns.countplot(y=col, data=df_final[df_final[col].isin(order_main)], order=order_main, hue=col, palette='viridis', legend=False, dodge=False)

        plt.title(f'Top {top_n} Categorías en {col}')
        plt.xlabel('Cantidad')
        plt.ylabel(col)

        if 'condition' in df_final.columns:
            plt.subplot(1, 2, 2)
            if pd.isna(order_main_temp.index).any():
                 df_top_n_condition = df_temp_countplot[df_temp_countplot[col].isin(order_main)]
                 sns.countplot(y=col, hue='condition', data=df_top_n_condition, order=order_main, palette='pastel')
            else:
                 df_top_n_condition = df_final[df_final[col].isin(order_main)]
                 sns.countplot(y=col, hue='condition', data=df_top_n_condition, order=order_main, palette='pastel')

            plt.title(f'{col} vs. Condition (Top {top_n})')
            plt.xlabel('Cantidad')
            plt.ylabel(col)
        plt.tight_layout()
        plt.show()
    else:
        print(f"Columna {col} no encontrada en df_final para EDA categórico.")
```

```python
print("\n--- 4. Análisis de Variables Booleanas ---")
boolean_cols_eda = df_final.select_dtypes(include='bool').columns.tolist()
for col in boolean_cols_eda:
    if col in df_final.columns:
        print(f"\nAnálisis de la columna booleana: {col}")
        print(df_final[col].value_counts(normalize=True, dropna=False) * 100)
        if 'condition' in df_final.columns:
            g = sns.catplot(x=col, col="condition",
                        data=df_final, kind="count",
                        height=4, aspect=.7, palette='pastel')
            g.fig.suptitle(f'{col} vs. Condition (faceted)', y=1.03)
            plt.show()
    else:
        print(f"Columna {col} no encontrada para EDA booleano.")
```

```python
print("\n--- Limpieza Final de Características (Pandas/Numpy) ---")

if 'condition_encoded' not in df_final.columns:
    print("Error Crítico: La columna 'condition_encoded' no existe. El preprocesamiento no puede continuar.")
    print("Asegúrese de que la celda 15 (Codificación de la Variable Objetivo) se haya ejecutado correctamente.")
    y = pd.Series(dtype='float64')
    X_features_cleaned = pd.DataFrame()
    X_processed_df = pd.DataFrame()

else:
    y = df_final['condition_encoded'].copy()
    X_features_cleaned = df_final.drop(columns=['condition', 'condition_encoded']).copy()
    print(f"Dimensiones de X_features_cleaned antes de la limpieza: {X_features_cleaned.shape}")

    numeric_features_clean = ['price', 'base_price', 'initial_quantity', 'sold_quantity', 'available_quantity']
    numeric_features_clean = [col for col in numeric_features_clean if col in X_features_cleaned.columns]

    categorical_features_all_clean = [
        'country', 'state', 'city', 'listing_type_id', 'buying_mode', 'shipping_mode',
        'status', 'listing_source', 'international_delivery_mode', 'id_category'
    ]
    if 'seller_id' in X_features_cleaned.columns:
        categorical_features_all_clean.append('seller_id')

    categorical_features_all_clean = [col for col in categorical_features_all_clean if col in X_features_cleaned.columns]

    boolean_features_clean = ['shipping_local_pick_up', 'free_shipping', 'accepts_mercadopago', 'automatic_relist']
    boolean_features_clean = [col for col in boolean_features_clean if col in X_features_cleaned.columns]

    text_features_to_keep_as_is_after_fillna = ['title', 'warranty', 'tags', 'shipping_methods', 'non_mercado_pago_payment_methods', 'date_created']
    text_features_to_keep_as_is_after_fillna = [col for col in text_features_to_keep_as_is_after_fillna if col in X_features_cleaned.columns]

    print(f"\nProcesando {len(numeric_features_clean)} características numéricas...")
    for col in numeric_features_clean:
        median_val = X_features_cleaned[col].median()
        X_features_cleaned[col].fillna(median_val, inplace=True)
        if col in ['price', 'base_price']:
            X_features_cleaned[col] = np.log1p(X_features_cleaned[col])

        mean_val = X_features_cleaned[col].mean()
        std_val = X_features_cleaned[col].std()
        if std_val == 0:
            X_features_cleaned[col] = 0
        else:
            X_features_cleaned[col] = (X_features_cleaned[col] - mean_val) / std_val

    print(f"\nProcesando {len(categorical_features_all_clean)} características categóricas...")
    temp_ohe_dfs = []

    cols_to_keep_directly = numeric_features_clean + boolean_features_clean + text_features_to_keep_as_is_after_fillna
    X_base_features = X_features_cleaned[cols_to_keep_directly].copy()

    for col in categorical_features_all_clean:
        X_features_cleaned[col] = X_features_cleaned[col].fillna('missing_placeholder_final').astype(str)

        if col in ['country', 'state', 'city', 'id_category', 'seller_id']:
            top_n_val = 10
            top_categories = X_features_cleaned[col].value_counts().nlargest(top_n_val).index
            X_features_cleaned[col] = X_features_cleaned[col].apply(lambda x: x if x in top_categories else 'Other')

        ohe_df = pd.get_dummies(X_features_cleaned[col], prefix=col, dummy_na=False)
        temp_ohe_dfs.append(ohe_df.reset_index(drop=True))

    X_features_cleaned = pd.concat([X_base_features.reset_index(drop=True)] + temp_ohe_dfs, axis=1)

    print(f"\nProcesando {len(boolean_features_clean)} características booleanas...")
    for col in boolean_features_clean:
        mode_val = X_features_cleaned[col].mode()
        if not mode_val.empty:
            X_features_cleaned[col].fillna(mode_val[0], inplace=True)
        else:
            X_features_cleaned[col].fillna(False, inplace=True)
        X_features_cleaned[col] = X_features_cleaned[col].astype(int)

    print(f"\nProcesando {len(text_features_to_keep_as_is_after_fillna)} características de texto (solo imputación)...") # CORRECCIÓN AQUÍ
    for col in text_features_to_keep_as_is_after_fillna:
        fill_value = ''
        if col == 'warranty':
            fill_value = 'Sin_Garantia_o_Nulo_Final'
        elif col == 'tags':
            fill_value = 'Sin_Tags_o_Nulo_Final'
        elif col == 'date_created':
             fill_value = 'missing_date_final'
        X_features_cleaned[col].fillna(fill_value, inplace=True)

    X_processed_df = X_features_cleaned
    print(f"\nDimensiones de X_processed_df después del preprocesamiento: {X_processed_df.shape}")
    print("\nPrimeras filas de X_processed_df:")
    print(X_processed_df.head())
```
